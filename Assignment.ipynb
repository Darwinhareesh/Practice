```python
import plotly.express as px
import plotly.graph_objects as go

# Bar chart for Rented Bike Count vs Hour
fig_bar = px.bar(data, x='Hour', y='Rented Bike Count', title='Rented Bike Count vs Hour',color='Rented Bike Count')
fig_bar.show()
```

```python
fig_temp = px.bar(data, x='Rented Bike Count', y='Holiday', title='Temperature vs Hour',color='Holiday')
fig_temp.show()
```

```python
# Line chart for Temperature vs Hour
fig_temp = px.line(data, x='Hour', y='Temperature(°C)', title='Temperature vs Hour',color='Temperature(°C)')
fig_temp.show()
```

```python
# Line chart for Humidity vs Hour
fig_humidity = px.line(data, x='Hour', y='Humidity(%)', title='Humidity vs Hour',color='Humidity(%)')
fig_humidity.show()
```

```python
# Line chart for Wind speed vs Hour
fig_wind = px.line(data, x='Hour', y='Wind speed (m/s)', title='Wind speed vs Hour',color='Wind speed (m/s)')
fig_wind.show()
```

```python

# Line chart for Dew point temperature vs Hour
fig_dew_point = px.line(data, x='Hour', y='Dew point temperature(°C)', title='Dew point temperature vs Hour',color='Dew point temperature(°C)')
fig_dew_point.show()
```

```python
# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

# Extract month from the Date column
data['Month'] = data['Date'].dt.month
```

```python
# Aggregate data by Month to get total bike rentals per month
df_monthly = data.groupby('Month')['Rented Bike Count'].sum().reset_index()
```

```python
# Bar chart for Rented Bike Count vs Month
fig_monthly = px.bar(df_monthly, x='Month', y='Rented Bike Count', title='Rented Bike Count vs Month')
fig_monthly.show()
```
